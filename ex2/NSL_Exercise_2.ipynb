### Exercise 02.1
- Compute the following 1D integral via Monte Carlo 
$$I = \int_0^1 \frac{\pi}{2}\cos(\pi x/2) dx = 1$$

1. Sampling a uniform distribution in $[0,1]$
2. Using importance sampling (i.e. sampling a non-uniform probability in $[0,1]$)

```python
# Include libs
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats
```

```python
xu, yu, erru = np.loadtxt("2.1/output/uniform.out", usecols=(0,1,2), delimiter=' ', unpack='true')
xi, yi, erri = np.loadtxt("2.1/output/importance.out", usecols=(0,1,2), delimiter=' ', unpack='true')

plt.figure(figsize=(15,5))

plt.errorbar(xu,yu,yerr=erru, label = 'Uniform sampling')
plt.errorbar(xi,yi,yerr=erri, label = 'Importance sampling')
plt.legend()
plt.hlines(1, 0, 100, colors='r', linestyles='dashed')
plt.xlabel('Block number')
plt.ylabel('I')
plt.grid(True)

plt.show()
```
